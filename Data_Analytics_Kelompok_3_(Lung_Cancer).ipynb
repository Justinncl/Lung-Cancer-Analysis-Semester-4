(example: 1 becomes 8, 8 becomes 1)

```python
contributing_factors = ['Air Pollution', 'Alcohol use', 'Dust Allergy', 'OccuPational Hazards', 'Genetic Risk', 'chronic Lung Disease', 'Balanced Diet', 'Obesity', 'Smoking', 'Passive Smoker']

for col in contributing_factors:
    if col in data.columns:
        data[col] = 9 - data[col]

display(data.head(25))
```

Displaying mean age and most frequent gender affected at each level of cancer (low, medium, high) & Displaying most contributing factors to lung cancer in each level

```python
  level_analysis = data.groupby('Level').agg(
      mean_age=('Age', 'mean'),
      most_frequent_gender=('Gender', lambda x: x.mode()[0])
  )

  display(level_analysis)

  print()
  print()

  contributing_factors = ['Air Pollution', 'Alcohol use', 'Dust Allergy', 'OccuPational Hazards', 'Genetic Risk', 'chronic Lung Disease', 'Balanced Diet', 'Obesity', 'Smoking', 'Passive Smoker']

  ranked_factors_by_level = {}


  for level in data['Level'].unique():
      data_level = data[data['Level'] == level]
      factor_means = data_level[contributing_factors].mean()

      sorted_factors = factor_means.sort_values(ascending=True)

      ranked_factors_by_level[level] = sorted_factors


  print("Summary of Ranked Contributing Factors by Cancer Level:")
  for level, ranked_factors in ranked_factors_by_level.items():
      print(f"\nLevel: {level}")
      print(ranked_factors)
```

Visualization of Contributing Factors to Lung Cancer based on Lung Cancer Level

```python
for level, ranked_factors in ranked_factors_by_level.items():
    plt.figure(figsize=(10, 6))
    sns.barplot(x=ranked_factors.index, y=ranked_factors.values, palette='viridis')
    plt.title(f'Ranked Contributing Factors for {level} Cancer Level')
    plt.xlabel('Contributing Factor')
    plt.ylabel('Mean Risk Contribution ')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
